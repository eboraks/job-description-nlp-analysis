```python
import wikipedia
import requests
import spacy
from collections import Counter
import matplotlib.pyplot as plt
import spacy
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
result = wikipedia.search("USA")
page_name = result[0]
print("First result {}".format(page_name))
page = wikipedia.page(page_name, preload= True, auto_suggest=False)
print(page.url)
```

```python
nlp = spacy.load('en_core_web_lg')
doc = nlp(page.content)
persons_vc_map  = {ent.text:ent.vector  for ent in doc.ents if ent.label_=='PERSON' }
persons_vc_map.keys()
```

```python
persons_vc_map.items()
```

```python
df = pd.DataFrame([(key , val ) for key , val in persons_vc_map.items()])
X = np.array([a for a in df[1]])
X[2]
```

```python
kmeans_persons = KMeans(n_clusters=20).fit(X)
```

```python
len(kmeans_persons.labels_)
```

```python
df ['labels'] =  kmeans_persons.labels_
```

```python
for label, names in df.groupby('labels'):
    print("cluster {} , values : {}".format( label, names[0].tolist()))
    print("++++++++++++++++")
```
