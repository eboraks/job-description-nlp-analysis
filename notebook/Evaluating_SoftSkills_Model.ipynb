```python
import spacy
import os 
import pandas as pd
from spacy import displacy

```

```python
nlp = spacy.load("../output/model-best")
```

```python
lnkjobs = pd.read_csv("../data/dice_data_scientist_2021-12-14T17:42:37.csv")
```

```python
results = [x for x in lnkjobs['job_desc']]
len(results)
```

```python
for job_desc in results:
    print('-----------------------------------------------------------------')
    doc = nlp(job_desc)
    displacy.render(doc, style='ent', jupyter=True)
```

```python
doc2 = nlp("Proficient programming experience using SAS Python R.")

displacy.render(doc2, style='dep', jupyter=True)

for t in doc2:
    print((t.text, t.dep_))
```
